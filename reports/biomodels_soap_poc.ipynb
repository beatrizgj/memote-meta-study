```python
from lxml import html
import requests
```

```python
response = requests.get("https://www.ebi.ac.uk/biomodels-main/path2models?cat=genome-scale")
response.raise_for_status()
```

```python
tree = html.fromstring(response.content)
```

```python
model_ids = tree.xpath("//li/a[contains(@title,'Access to this model')]/@href")
```
